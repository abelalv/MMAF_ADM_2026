# 1. Predicción del precio del Bitcoin

## 1.1 Contexto

Las criptomonedas nacieron en 2009 con la llegada de Bitcoin, creada por alguien que se hacía llamar **Satoshi Nakamoto**. Todo empezó como una idea para tener dinero digital que no dependiera de bancos ni gobiernos. Para ello se usó una tecnología llamada **blockchain**. la cual es una forma de registrar y compartir información en una red de muchas computadoras simultáneamente y no se pueden modificar fácilmente, lo que la hace muy segura y transparente. Desde entonces, han aparecido miles de criptomonedas con diferentes propósitos y comunidades detrás. En el 2010 se realizó la primera transacción comercial con Bitcoin, cuando un programador llamado Laszlo Hanyecz pagó 10,000 BTC por dos pizzas. con un precio de apenas unos pocos centavos por Bitcoin en ese momento. Ahora, más de una década después, Bitcoin ha alcanzado precios de decenas de miles de dólares por unidad y ha capturado la atención de inversores, instituciones financieras y gobiernos en todo el mundo. Fuente: [Historia de Bitcoin](https://www.investopedia.com/articles/forex/121815/bitcoins-price-history.asp).
