## Una breve historia del contenedor marítimo

El contenedor marítimo, también llamado "container" (término inglés), se remonta a la Segunda Guerra Mundial. Era utilizado para el transporte seguro de materiales bélicos.

Fue creado por el estadounidense Malcolm McLean, un camionero que nació el 1913 en una familia de clase media de Carolina del Norte. Fue en 1953, cuando este empresario observó que casi todos los puertos de la costa oeste estaban completamente saturados. Para optimizar el tiempo decidió cargar directamente los remolques de sus camiones a los barcos.

Partiendo de este modelo operativo también se percató que suponía un ahorro de tiempo, pero no era demasiado rentable, ya que el espacio utilizado por los remolques era demasiado grande. A partir de ahí, le surge la idea de quitar el chasis y embarcar solo la parte del remolque, es decir, la "caja" en sí. Y así fue como inventó el contenedor que conocemos hoy en día.

En menos de diez años, el uso de los contenedores de transporte se extendió por todo el mundo, ya que permitió optimizar el espacio a bordo y ahorrar tiempo para las operaciones de manipulación. Fue en 1961 cuando los containers de McLean se modifican para implementar una estandarización internacional.
A partir de ese momento, se definieron las dimensiones tipo aprobadas por la Organización Internacional de Normalización (ISO): 20, 30 y 40 pies, estableciendo así las medidas que aún se utilizan como referencia en el transporte marítimo global.

Hoy en día, los contenedores están hechos mayoritariamente de acero, aunque también podemos encontrarlos de aluminio. Además, existe una gran variedad de contenedores para compensar la escasez y optimizar el transporte de mercancías.

Es importante saber que los contenedores cumplen con las normas internacionales ISO que rigen el transporte de mercancías. De esta manera las dimensiones en milímetros (mm) son las siguientes:

•	Dry 20 pies: 6 058 mm (longitud), 2 438 mm (ancho) y 2 591 mm (altura).

•	High Cube 20 pies: 6 058 mm (longitud), 2 438 mm (ancho) y 2 896 mm (altura).

•	Dry 40 pies: 12 192 mm (longitud), 2 438 mm (ancho) y 2 591 mm (altura).

•	High Cube 40 pies: 12 192 mm (longitud), 2 438 mm (ancho) y 2 698 mm (altura).
